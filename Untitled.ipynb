```python
import pymongo
import pandas as pd
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)

#extract to DB
db = client.gggDB
emo_collections = db.emotion_db.find()
df = pd.DataFrame(list(emo_collections))
```

```python
df
```

```python
count_emo = df.groupby("Emotion").count()
```

```python
data_df = count_emo.reset_index()
```
